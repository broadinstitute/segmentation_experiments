# Experiment with image cropping

```python
from skimage import io
import numpy as np
```

```python
sample = io.imread(purple_img_paths[5])[:, :, [0, 1, 2]]
io.imshow(sample)
sample.shape
```

```python
cropped = sample[:64, :64, :]
io.imshow(cropped)
cropped.shape
```

```python
row_start = np.random.randint(0, sample.shape[0] - 64)
col_start = np.random.randint(0, sample.shape[1] - 64)

row_end = row_start + 64
col_end = col_start + 64

rand_crop = sample[row_start:row_end, col_start:col_end, :]
io.imshow(rand_crop)
```
