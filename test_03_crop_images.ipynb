# Experiment with image cropping

```python
from skimage.util import crop
from skimage import io
import numpy as np
```

```python
sample = io.imread(purple_img_paths[5])[:, :, [0, 1, 2]]
io.imshow(sample)
sample.shape
```

```python
cropped = sample[:64, :64, :]
io.imshow(cropped)
cropped.shape
```

```python
rand_start = np.random.randint(0, min(sample.shape[0]-64, sample.shape[1]-64))
end = rand_start + 64

rand_crop = sample[rand_start:end, rand_start:end, :]
io.imshow(rand_crop)
print(rand_start)
```
